```python
def remove_consecutive(string):
    result = ""
    count = 0
    prev_char = None

    for char in string:
        if char == prev_char:
            count += 1
        else:
            count = 1

        if count <= 3:
            result += char

        prev_char = char

    return result


# Example usage
s = "tuuuuuriiiiiing"
output = remove_consecutive(s)
print(output)  # Output: tuuuriiing
```
